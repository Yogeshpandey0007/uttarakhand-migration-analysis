```python
import pandas as pd

sheet_id = "1zRLRPaL08Ehsp_AFGA68b-JZJ7CuVPdYuBjKl37U7Ic"

url = f"https://docs.google.com/spreadsheets/d/{sheet_id}/export?format=xlsx"

# Read all sheets
all_sheets = pd.read_excel(
    url,
    sheet_name=None
)
```

```python
all_sheets.keys()
```

```python
for name, df in all_sheets.items():

    # Remove empty rows
    df.dropna(how='all', inplace=True)

    # Remove empty columns
    df.dropna(axis=1, how='all', inplace=True)

    # Remove extra spaces from column names
    df.columns = df.columns.str.strip()

    # Reset index
    df.reset_index(drop=True, inplace=True)

    # Remove duplicate rows
    df.drop_duplicates(inplace=True)

    # Save cleaned sheet
    all_sheets[name] = df

print("Cleaning completed.")
```

```python
migration_df = migration_df[
    migration_df['District'] != 'Total Uttarakhand'
]

migration_df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(16,7))

ax = sns.barplot(
    x='District',
    y='Temporary Migrants',
    data=migration_df
)

# Add values above bars
for p in ax.patches:

    ax.annotate(
        f'{p.get_height():,.0f}',
        (p.get_x() + p.get_width()/2., p.get_height()),
        ha='center',
        va='bottom',
        fontsize=10
    )

plt.xticks(rotation=90)

plt.title(
    'District-wise Temporary Migration in Uttarakhand (2008–2018)',
    fontsize=18,
    fontweight='bold'
)

plt.xlabel(
    'Districts',
    fontsize=12
)

plt.ylabel(
    'Number of Temporary Migrants',
    fontsize=12
)

plt.tight_layout()

plt.show()
```

## Key Insights

- Tehri Garhwal recorded the highest temporary migration with more than 71,000 migrants between 2008–2018.

- Almora and Pauri Garhwal also show significantly high temporary migration, indicating strong outward movement from hill districts.

- Plains districts such as Haridwar and Udham Singh Nagar show comparatively lower migration levels.

- The analysis suggests that employment opportunities, educational migration, and economic challenges are major drivers of migration in mountainous regions of Uttarakhand.

- Hill districts appear more vulnerable to temporary migration compared to urban and industrial districts.

```python
plt.figure(figsize=(14,6))

x = range(len(migration_df))

plt.bar(
    x,
    migration_df['Temporary Migrants'],
    label='Temporary Migration'
)

plt.bar(
    x,
    migration_df['Permanent Migrants'],
    label='Permanent Migration'
)

plt.xticks(
    x,
    migration_df['District'],
    rotation=90
)

plt.title('Temporary vs Permanent Migration')

plt.legend()

plt.show()
```

## Key Insights

- Temporary migration is significantly higher than permanent migration across almost all districts of Uttarakhand.

- Tehri Garhwal recorded the highest temporary migration, indicating a large working population moving outside the district for seasonal or employment-related opportunities.

- Pauri Garhwal shows one of the highest permanent migration figures, suggesting long-term relocation of residents from rural hill areas.

- Almora, Chamoli, and Pithoragarh also demonstrate substantial migration pressure, highlighting economic and employment challenges in hill districts.

- Plains districts such as Haridwar and Udham Singh Nagar show comparatively lower migration levels, likely due to better employment opportunities and urban infrastructure.

- The gap between temporary and permanent migration indicates that many migrants initially leave temporarily for employment or education, but over time a significant portion may permanently relocate.

- The analysis reflects growing demographic imbalance and rural depopulation trends in Uttarakhand’s mountainous regions.

```python
merged_df = migration_df.merge(
    reasons_df,
    on='District'
)

merged_df = merged_df.merge(
    occupation_df,
    on='District'
)
```

```python
merged_df.head(13)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(16,10))

sns.heatmap(
    merged_df.select_dtypes(include='number').corr(),
    annot=True,
    cmap='coolwarm'
)

plt.title(
    'Correlation Between Migration, Employment and Rural Economy Indicators',
    fontsize=16,
    fontweight='bold'
)

plt.show()
```

## Key Insights

- The correlation heatmap reveals strong relationships between migration patterns, employment conditions, rural economy indicators, and livelihood dependency across Uttarakhand districts.

- Temporary migration and permanent migration show a very strong positive correlation, indicating that districts with high temporary migration often experience long-term population outflow over time.

- MGNREGA dependency shows a strong positive relationship with migration indicators, suggesting that economically vulnerable districts rely more heavily on government-supported employment schemes.

- Agriculture dependency demonstrates negative correlations with several migration indicators, reflecting the declining sustainability of traditional rural livelihoods in some districts.

- Employment-driven migration shows strong relationships with education migration and infrastructure-related factors, highlighting the importance of economic and developmental opportunities in migration decisions.

- Health, education, and infrastructure variables also exhibit interconnected relationships, indicating that migration is influenced by multiple socio-economic factors rather than a single cause.

- The heatmap confirms that migration in Uttarakhand is closely linked with rural economic stress, employment limitations, infrastructure challenges, and changing livelihood patterns.

- The analysis provides valuable insights for policymakers to identify vulnerable districts and design targeted rural development, employment generation, and migration management strategies.

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))

sns.scatterplot(
    x='Employment %',
    y='Permanent Migrants',
    data=merged_df,
    s=150
)

# Add district labels
for i in range(len(merged_df)):

    plt.text(
        merged_df['Employment %'].iloc[i],
        merged_df['Permanent Migrants'].iloc[i],
        merged_df['District'].iloc[i],
        fontsize=9
    )

plt.title(
    'Relationship Between Employment Migration and Permanent Migration',
    fontsize=16,
    fontweight='bold'
)

plt.xlabel('Employment Driven Migration (%)', fontsize=12)

plt.ylabel('Permanent Migrants', fontsize=12)

plt.show()
```

## Key Insights

- The scatter plot shows the relationship between employment-driven migration and permanent migration across Uttarakhand districts.

- Pauri Garhwal and Tehri Garhwal appear in the upper-middle region of the graph, indicating both high employment-driven migration and high permanent migration.

- Almora and Chamoli also show substantial permanent migration levels, suggesting that employment-related migration often becomes long-term migration in hill districts.

- Haridwar and Udham Singh Nagar display high employment migration percentages but comparatively lower permanent migration levels, likely due to stronger local economic opportunities and industrial development.

- Districts such as Uttarkashi and Bageshwar show lower employment migration percentages along with lower permanent migration, although migration pressure still exists.

- The overall trend suggests that districts with weaker local employment opportunities are more vulnerable to long-term population outflow.

- The analysis highlights how employment conditions strongly influence demographic change and rural depopulation in Uttarakhand.

```python
plt.figure(figsize=(10,6))

sns.scatterplot(
    x='Agriculture/Horticulture/Animal Husbandry %',
    y='Permanent Migrants',
    size='MGNREGA %',
    data=merged_df,
    sizes=(100,1000)
)

# Add district labels
for i in range(len(merged_df)):

    plt.text(
        merged_df['Agriculture/Horticulture/Animal Husbandry %'].iloc[i],
        merged_df['Permanent Migrants'].iloc[i],
        merged_df['District'].iloc[i],
        fontsize=9
    )

plt.title(
    'Agriculture Dependency vs Permanent Migration in Uttarakhand',
    fontsize=16,
    fontweight='bold'
)

plt.xlabel(
    'Agriculture Dependency (%)',
    fontsize=12
)

plt.ylabel(
    'Permanent Migrants',
    fontsize=12
)

plt.show()
```

## Key Insights

- The bubble chart illustrates the relationship between agriculture dependency, permanent migration, and MGNREGA dependency across Uttarakhand districts.

- Pauri Garhwal stands out with very high permanent migration despite relatively lower agriculture dependency, indicating severe long-term outmigration and economic stress.

- Tehri Garhwal, Almora, and Chamoli also show high permanent migration levels along with moderate agricultural dependence, suggesting limited sustainability of traditional rural livelihoods.

- Larger bubble sizes in districts such as Pauri Garhwal and Rudraprayag indicate stronger dependence on MGNREGA employment schemes.

- Haridwar and Udham Singh Nagar exhibit high agriculture dependency but lower permanent migration levels, likely due to better economic opportunities and urban-industrial development.

- Hill districts generally show a combination of higher migration pressure and greater dependence on government-supported employment schemes.

- The analysis suggests that agriculture alone is insufficient to sustain livelihoods in many rural districts, contributing to migration and demographic imbalance.

- The findings emphasize the importance of rural economic diversification, local employment generation, and modernization of agriculture to reduce long-term migration pressure in Uttarakhand.

# Conclusion

This project analyzed migration patterns, rural occupation structures, agricultural dependency, MGNREGA reliance, and uninhabited villages across districts of Uttarakhand using multi-year government survey datasets.

The analysis revealed that hill districts such as Pauri Garhwal, Tehri Garhwal, Almora, and Chamoli experience significantly higher migration pressure compared to plains districts. Employment-related migration emerged as one of the strongest drivers of population movement, while education migration also contributes substantially to long-term demographic change.

The study further identified a strong relationship between rural economic vulnerability, dependence on agriculture, and increasing reliance on government-supported employment schemes such as MGNREGA. Districts with higher migration levels also tend to show larger numbers of uninhabited villages, indicating growing rural depopulation and weakening local economies.

Correlation and scatter plot analysis demonstrated that migration in Uttarakhand is influenced by interconnected socio-economic factors including employment opportunities, infrastructure, education access, agricultural sustainability, and livelihood diversification.

Overall, the project highlights the urgent need for:
- rural employment generation
- economic diversification
- infrastructure development
- agricultural modernization
- education and skill development initiatives

to reduce migration pressure and promote sustainable rural development in Uttarakhand.
